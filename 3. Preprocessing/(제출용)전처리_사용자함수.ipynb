cleanName 함수
- [오늘 종료!] 레진아트를 이용한 휴대폰 케이스 만들기
- 강좌명이 이런 식으로 신청 종료 날짜를 알려 주는데, 이런 대괄호로 감싸져 있는 글자를 지우는 함수이다.

```python
def cleanName(name):
  name = re.sub('\[{1}.+\] ','',name)
  name = name.strip()
  return name
```

stemming 함수
- Okt를 이용해 원형 형태소로 댓글 내용을 정제해 준다.
- 형태소의 원형으로 댓글을 정제하면 감성사전 search가 용이해진다.

```python
def stemming(review):
  malist5 = okt.pos(review, norm=False, stem=True)

  comment = ''
  last = ''
  for pos in malist5:
    if pos[1] in ['Verb','Punctuation','Adverb']:
      comment = comment + pos[0] + ' '
    elif last == pos[1]:
      comment = comment + ' ' + pos[0]
    elif pos[1] == 'Josa':
      comment = comment + pos[0] + ' '
    else:
      comment = comment + pos[0]
    last = pos[1]
  return comment

class101['댓글내용'] = class101.apply(lambda x: stemming(str(x['댓글내용'])),axis=1)
class101
```

countValue 함수
- list 안에 특정 요소가 몇 개 존재하는지 dictionary 형태로 반환해 주는 사용자 함수
- 이 함수를 이용해 감성분석 결과를 시각화함

```python
def countValue(list_):
  countDict = {}
  values = set(list_)
  values = list(values)
  for value in values:
    num = list_.count(value)
    countDict[value] = num
    
  return countDict
```
